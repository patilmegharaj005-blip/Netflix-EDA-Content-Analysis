```python
df = pd.read_csv('netflix_titles.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not Available', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.head()
```

```python
df['type'].value_counts()
```

```python
df['type'].value_counts().plot(kind='bar')
plt.title('Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries')
plt.show()
```

```python
df['release_year'].value_counts().sort_index().tail(10)
```

```python
df['release_year'].value_counts().sort_index().plot()
plt.title('Content Release Trend')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
df['rating'].value_counts().head(10)
```

```python
df.to_csv('cleaned_netflix_data.csv', index=False)
```

```python
# Movies vs Release Year trend
df.groupby('release_year')['type'].count().plot()
```

```python
# Content type by country
df.groupby(['country','type']).size().unstack().head()
```

```python
print("Insight: Netflix has significantly more Movies than TV Shows, indicating higher demand for movie content.")
```

```python
movies = df[df['type'] == 'Movie']
tv = df[df['type'] == 'TV Show']

movies.groupby('release_year').size().plot(label='Movies')
tv.groupby('release_year').size().plot(label='TV Shows')

plt.legend()
plt.title("Growth of Movies vs TV Shows Over Time")
plt.show()
```

```python
from collections import Counter

genres = df['listed_in'].str.split(', ').explode()
top_genres = genres.value_counts().head(10)

top_genres.plot(kind='bar')
plt.title("Top Genres on Netflix")
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot()
plt.title("Content Added to Netflix Over Time")
plt.show()
```
